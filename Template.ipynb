# Config


```python
num_workers = 8
batch_size = 512
memory_bank_size = 4096
seed = 1
max_epochs = 1
input_size=32
pl.seed_everything(seed)
path_to_train = "../Datasets/cifar10/train/"
path_to_test = "../Datasets/cifar10/test/"
```

# Load Data

```python
# Augmentations typically used to train on cifar-10
finetuning_transforms = torchvision.transforms.Compose(
    [
        torchvision.transforms.RandomCrop(input_size, padding=4),
        torchvision.transforms.RandomHorizontalFlip(),
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize(
            mean=collate.imagenet_normalize["mean"],
            std=collate.imagenet_normalize["std"],
        ),
    ]
)

# No additional augmentations for the test set
test_transforms = torchvision.transforms.Compose(
    [
        torchvision.transforms.Resize((input_size, input_size)),
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize(
            mean=collate.imagenet_normalize["mean"],
            std=collate.imagenet_normalize["std"],
        ),
    ]
)
 
```

```python

# We use the moco augmentations for training moco
dataset_train = LightlyDataset(input_dir=path_to_train)

# Since we also train a linear classifier on the pre-trained moco model we
# reuse the test augmentations here (MoCo augmentations are very strong and
# usually reduce accuracy of models which are not used for contrastive learning.
# Our linear layer will be trained using cross entropy loss and labels provided
# by the dataset. Therefore we chose light augmentations.)
dataset_finetuning = LightlyDataset(
    input_dir=path_to_train, transform=finetuning_transforms
)

dataset_test = LightlyDataset(input_dir=path_to_test, transform=test_transforms)
```

```python
# MoCo v2 uses SimCLR augmentations, additionally, disable blur
collate_fn = SimCLRCollateFunction(
    input_size=input_size,
    gaussian_blur=0.0,
)

dataloader_train = torch.utils.data.DataLoader(
    dataset_train,
    batch_size=batch_size,
    shuffle=True,
    collate_fn=collate_fn,
    drop_last=True,
    num_workers=num_workers,
)

dataloader_finetuning = torch.utils.data.DataLoader(
    dataset_finetuning,
    batch_size=32,  # Change batch size to 32
    shuffle=True,
    drop_last=True,
    num_workers=num_workers,
)

dataloader_test = torch.utils.data.DataLoader(
    dataset_test,
    batch_size=32,
    shuffle=False,
    drop_last=False,
    num_workers=num_workers,
)
```

```python
import torch
from tqdm import tqdm
import numpy as np

def dataloader2array(dataloader,device):
    """
    This function extracts features and labels from the provided dataloader using the given model.

    Parameters:
    dataloader (DataLoader): The DataLoader object to iterate over.
    device (torch.device): The device (CPU or GPU) where the computations will be performed.

    Returns:
    X (np.ndarray): The array of features.
    y (np.ndarray): The array of labels.
    fname (np.ndarray): The array of filenames.
    """
    
    # Initialize features and labels as empty lists
    X = []
    y = []
    fname = []

    for images, labels, fnames in tqdm(dataloader):
        with torch.no_grad():
            features = images.cpu().numpy()
            labels = labels.cpu().numpy()
            fnames = np.array(fnames)

            X.append(features)
            y.append(labels)
            fname.append(fnames)

    # Convert lists to numpy arrays
    X = np.concatenate(X, axis=0)
    y = np.concatenate(y, axis=0)
    fname = np.concatenate(fname, axis=0)
    return X, y, fname
X_test, y_test,fname_test= dataloader2array(dataloader_test,device)
```
